```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the enriched supermarket data
df = pd.read_csv("supermarkets_data_enriched.csv")
```

c) 1

```python
brand_counts = df['brand'].value_counts()
print(brand_counts)
```

c) 2

```python
brand_counts.plot(kind='bar', figsize=(20,4))
plt.title("Supermarkets per Brand")
plt.ylabel("Count")
plt.xlabel("Brand")
plt.show()
```

c) 3

```python
grid = sns.PairGrid(df[['lat',
                     'lon',
                     'pop',
                     'pop_dens',
                     'frg_pct',
                     'emp']],
                 height=1.2,
                 aspect=1)
grid.map_upper(sns.scatterplot,color='darkgreen', s=10)
grid.map_lower(sns.scatterplot, color='darkgreen', s=10)
grid.map_diag(plt.hist, color='brown')
```

```python
df_sub = df.loc[df['brand'].isin(['Denner', 'Volg', 'Landi', 'Coop', 'Migros'])]

colors = {
    'Denner': 'red',
    'Volg': 'blue',
    'Landi': 'green',
    'Coop': 'orange',
    'Migros': 'brown'
}

plt.scatter(df_sub['lon'], df_sub['lat'], c=df_sub['brand'].map(colors))
```
